# Pre-processing

# Case Folding

```python
df['message to examine'][0].lower()
df['message to examine']=df[  'message to examine'].str.lower()
df.head()
```

# Removing Tags

```python
import re
def remove_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'',text)

df['message to examine']=df['message to examine'].apply(remove_tags)
df.head()
```

# Removing URL

```python
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)

df['message to examine']=df['message to examine'].apply(remove_url)
df.head()
```

# Removing Punctuations

```python
import string
def remove_punctuations(text):
  translator = str.maketrans('', '', string.punctuation)
  return text.translate(translator)

df['message to examine']=df['message to examine'].apply(remove_punctuations)
df.head()
```

# Replace Short Hands

```python
from google.colab import files

uploaded = files.upload()

file_path = 'slang.txt'

data_dict = {}
with open(file_path, 'r') as file:
    for line in file:
        line = line.strip()
        if line and '=' in line:
            key, value = line.split('=', 1)
            data_dict[key] = value

print(data_dict)

def chat_conversion(text):
    new_text =[]
    for c in text.split():
        if c.upper() in data_dict:
            new_text.append(data_dict[c.upper()])
        else:
            new_text.append(c)
    return " ".join(new_text)
```

```python
df['message to examine']=df['message to examine'].apply(chat_conversion)
df.head()
```

# Stop words removal

```python
import nltk
from nltk.corpus import stopwords
# nltk.download('stopwords')
stopwords.words('english')

def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)

remove_stopwords(df['message to examine'][6])
```
